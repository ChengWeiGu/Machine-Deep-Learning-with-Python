```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# to generate the scattering points
x = np.linspace(0,10,100)
noise = np.random.randn(100)*5
y = 1 + 3*x + noise

plt.scatter(x,y)
```

```python
# LaTeX in Jupyter Notebooks: http://data-blog.udacity.com/posts/2016/10/latex-primer/
# the cost function J(θ0,θ1) based on the hypothesis hθ(x) = θ0 + θ1*x

contour_X, contour_Y = np.meshgrid(np.linspace(-0.5,2,20),np.linspace(2.5,3.5,20))
pairs = np.c_[np.ravel(contour_X),np.ravel(contour_Y)]
#θ0, θ1 = np.ravel(pairs[:,0]), np.ravel(pairs[:,1])

J = np.array([])
for index, θ in enumerate(pairs):
    θ0, θ1 = θ[0], θ[1]
    hθ = θ0 + θ1*x
    J_element = (1/2/len(x))*np.sum((hθ-y)**2)
    J = np.append(J,J_element)
```

```python
# plot the conotur diagram
plt.contourf(contour_X,contour_Y, J.reshape(20,20),10, alpha = 0.8,cmap = plt.cm.hot)
C = plt.contour(contour_X,contour_Y, J.reshape(20,20),10, colors = 'black', linewidth = .5)
plt.clabel(C, inline = True, fontsize = 10)
plt.colorbar()
plt.xlabel('θ0')
plt.ylabel('θ1')
plt.show()
```
